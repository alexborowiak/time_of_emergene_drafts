```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from matplotlib.patches import Rectangle
```

```python
def setup_toe_lineplot(ax, xupper=10):
    ax.grid(linestyle=':')
    ax.set_ylim(1900, 2100)
    ax.set_yticks(np.arange(1900, 2110, 10));
    ax.axhline(2020, color='k')
    ax.set_xlim(0, xupper);
    ax.set_xticks(np.arange(xupper));
```

```python
fig, (ax1, ax2)  = plt.subplots(1, 2,figsize=(8, 12))
setup_toe_lineplot(ax1)
setup_toe_lineplot(ax2)
ax1.set_title('TAS')
ax2.set_title('PR')

labels = ax1.get_xticklabels()
labels[0] = ''

rect1 = Rectangle((0.75, 2030), 0.5, 30, color='orange', label='Tropics')
ax1.add_patch(rect1)
labels[1] = 'Hawkins and Sutton 2012'

rect1 = Rectangle((1.75, 2060), 0.5, 40, color='green', label='Mid-Latitudes')
ax1.add_patch(rect1)
labels[2] = 'Hawkins and Sutton 2012'


rect1 = Rectangle((2.75, 1920), 0.5, 40, color='orange')
ax1.add_patch(rect1)
labels[3] = 'King et al. 2015'


rect1 = Rectangle((3.75, 1960), 0.5, 40, color='green')
ax1.add_patch(rect1)
labels[4] = 'King et al. 2015'


rect1 = Rectangle((4.75, 1980), 0.5, 40, color='blue', label='Arctic')
ax1.add_patch(rect1)
labels[5] = 'King et al. 2015'

ax1.set_xticklabels(labels, rotation=90);
ax1.legend()
```
